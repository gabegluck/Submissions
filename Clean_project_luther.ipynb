```python
import matplotlib.pyplot as plt
%matplotlib inline
import pickle
import random
import numpy as np
import pandas as pd
pd.options.display.mpl_style = 'default'
import statsmodels.formula.api as sm
from sklearn.linear_model import Ridge
```

```python
#import movies
with open("top500actors_top50directors.pkl", 'r') as picklefile:
    movie_dicts=pickle.load(picklefile)
#create training and test sets
random.shuffle(movie_dicts)
training_set=movie_dicts[:6000]
test_set=movie_dicts[6000:]
```

```python
#Making Simple Model
dataframe_all['Ones'] = 1.0
X = dataframe_all[['critic_rating', 'Ones']]
Y = dataframe_all['imdbRating']
linmodel = sm.OLS(Y, X).fit()
linmodel.summary()
```

```python
#make a list of features
with open("sum_of_actor_appearances_movies.pkl" , 'r') as picklefile:
    actor_movies=pickle.load(picklefile)
with open("sum_of_director_appearances_movies.pkl" , 'r') as picklefile:
    director_movies=pickle.load(picklefile)
#make actor features
actor_features=[]
for actor in sorted(actor_movies, key=actor_movies.get, reverse=True)[:500]:
    actor_features.append(actor)
actor_features.remove("N/A")
actor_features.remove("Kevin Kline")
print "Number of Actors: " + str(len(actor_features))
#make director features
director_features=[]
for director in sorted(director_movies, key=director_movies.get, reverse=True)[:50]:
    director_features.append(director)
print "Number of Directors: " + str(len(director_features))
#make genre features
genre_features=[]
for movie_dict in movie_dicts:
    for genre in movie_dict['Genre']:
        if genre not in genre_features:
            genre_features.append(genre)
print "number of genre features: " + str(len(genre_features))
features=list(genre_features+actor_features+director_features+ ['critic_rating'])
features.remove(None)
features.remove('David Gordon Green')
print "Total Number of Features: " + str(len(features))
```

```python
#make training data into numpy arrays
x_vectors=[]
for feature in features:
    vector=np.matrix([m[feature] for m in training_set]).transpose()
    x_vectors.append(vector)
x_input=np.column_stack(x_vectors)
y_true=np.matrix([m['imdbRating'] for m in training_set]).transpose()

#make test data into numpy arrays
x_vectors_test=[]
for feature in features:
    vector=np.matrix([m[feature] for m in test_set]).transpose()
    x_vectors_test.append(vector)
x_input_test=np.column_stack(x_vectors_test)
#y_true=np.matrix([m['imdbRating'] for m in test_set]).transpose()
```

```python
#look for best alpha for ridge regression
alphas=[i/2.0 for i in range(1,100)]
mse_list=[]
for my_alpha in alphas:
    clf = Ridge(alpha=my_alpha)
    clf.fit(x_input, y_true)
    predictions=clf.predict(x_input_test)
    mse=0
    for i in range(len(test_set)):
        mse+= abs(test_set[i]['imdbRating'] - predictions[i])
    mse=float(mse/float(len(test_set)))
    mse_list.append(mse)
#print alphas
#print mse_list
plt.scatter(alphas,mse_list)
plt.ylabel("MSE")
plt.xlabel("alpha")
```

```python
best_coef=alphas[mse_list.index(min(mse_list))]
print "The lowest mean error is %fl when the Ridge Regression cofficient is %.1f." % (min(mse_list),best_coef)

```

```python
x_vectors=[]
for feature in features:
    vector=np.matrix([m[feature] for m in movie_dicts]).transpose()
    x_vectors.append(vector)
x_input=np.column_stack(x_vectors)
y_true=np.matrix([m['imdbRating'] for m in movie_dicts]).transpose()
#Running best ridge Regression fit
clf = Ridge(alpha=best_coef)
clf.fit(x_input, y_true)
actuals=[]
predicted=[]
predictions=clf.predict(x_input_test)
for i in range(len(test_set)):
        actuals.append(test_set[i]['imdbRating'])
        predicted.append(predictions[i])
#make a dictionary of all the coefficients
coef_dict={}
for i in range(len(features)):
    coef_dict[features[i]]=clf.coef_[0][i]
#for coef in sorted(coef_dict, key=coef_dict.get, reverse=True):
    #print coef, coef_dict[coef]
#r Squared
#print clf.score(x_input,y_true)
```
