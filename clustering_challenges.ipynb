```python
#make model
vectorizer = TfidfVectorizer(stop_words='english',ngram_range=(1,1))
headline_vectors = vectorizer.fit_transform(all_headlines)
k_means = KMeans(init='k-means++', n_clusters=5, n_init=10)
k_means.fit(headline_vectors)
k_means.labels_
#make default dict with cluster number and headlines in each cluster
cluster_dict=defaultdict(list)
for index,cluster in enumerate(k_means.labels_):
    cluster_dict[cluster].append(all_headlines[index])
pp=PrettyPrinter()
pp.pprint(dict(cluster_dict))
#make model_dict
model_dict={}
model_dict['kmeans']=k_means
```

```python
#do same with Agglomerative and DBSCAN
dense_headline_vectors=headline_vectors.todense()
agglomerative_clustering=AgglomerativeClustering(affinity='euclidean', n_clusters=5)
print agglomerative_clustering.fit_predict(dense_headline_vectors)
#agglomerative_clustering.predict(dense_headline_vectors)
model_dict['agglomerative clustering']=agglomerative_clustering
```

#Challenge 6

Cluster the same data with MiniBatchKMeans. MiniBatchKMeans is a fast way to apply K-means to large data without much loss -- The results are very similar. Instead of using EVERY single point to find the new place of the centroid, MiniBatch just randomly samples a small number (like 100) in the cluster to calculate the new center. Since this is usually very close to the actual center, the algorithm gets there much faster. Try it and compare the results. Example on two-feature data2

```python
#get data
wholesale=pd.io.parsers.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00292/Wholesale%20customers%20data.csv')
wholesale=wholesale.astype(float)
#get rid of first two columns 
whole_sale_customers=wholesale.iloc[:,2:]
scaled_wholescale=scale(whole_sale_customers)
```

```python
plt.style.use('fivethirtyeight')
k_inertias=[]
for i in range(1,30):
    k_means = KMeans(n_clusters=i, n_init=10)
    k_means.fit(scaled_wholescale)
#     print k_means.inertia_,i
    k_inertias.append(k_means.inertia_)
#plt.plot(range(1,30),k_inertias, label='inertias' ) 
#look at differences
inertia_differences=abs(np.diff(k_inertias))

plt.plot(range(len(inertia_differences)), inertia_differences, label='inertia differences')
plt.legend()
```

```python
#decided on 5 clusters
k_means = KMeans(n_clusters=5, n_init=10)
k_means.fit(scaled_wholescale)
whole_sale_customers['cluster']=k_means.predict(scaled_wholescale)
```
